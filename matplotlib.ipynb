```python
#Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
#Matplotlib.

Matplotlib is a popular Python data visualization library used to create high-quality 2D and 3D plots, charts, and graphs. It provides a comprehensive set of tools for creating static, animated, and interactive visualizations.

Why is it used?

Matplotlib is widely used in various fields, including:

Data Science: To visualize and explore datasets, identify patterns, and communicate insights.
Scientific Computing: To visualize numerical results, such as simulations, experiments, and models.
Education: To create interactive and engaging visualizations for teaching and learning.
Research: To present research findings and results in a clear and concise manner.
Five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: plt.plot(): Used to create a line graph, where data points are connected by lines.
Scatter Plot: plt.scatter(): Used to create a scatter plot, where data points are represented as points on a 2D plane.
Bar Chart: plt.bar(): Used to create a bar chart, where data is represented as vertical or horizontal bars.
Histogram: plt.hist(): Used to create a histogram, which displays the distribution of data values.
Pie Chart: plt.pie(): Used to create a pie chart, which displays the proportion of each component in a dataset.
The Pyplot module is a part of Matplotlib that provides a MATLAB-like interface for creating plots. It's a convenient and easy-to-use way to create a wide range of plots and charts.




```

```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#plot a scatter plot.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)

plt.title("Scatter Plot of x and y")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
Q3. Why is the subplot() function used?

The subplot() function is used to create multiple plots within a single figure. It allows you to divide a figure into multiple subplots, each of which can contain a separate plot. This is useful when you want to compare multiple plots or show different aspects of the same data.

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title("Line 1")

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title("Line 2")

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title("Line 3")

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title("Line 4")

plt.tight_layout()
plt.show()
```

```python
Q4. What is a bar plot? Why is it used?

A bar plot, also known as a bar chart, is a type of plot used to visualize categorical data. It consists of rectangular bars with lengths proportional to the values they represent. The bars are typically arranged along a horizontal or vertical axis, with the categories represented on the other axis.

Bar plots are used to:

Compare categorical data: Bar plots are useful for comparing the values of different categories.
Show rankings: Bar plots can be used to show the ranking of categories based on their values.
Identify patterns: Bar plots can help identify patterns or trends in categorical data.
Plotting a bar plot and a horizontal bar plot:

Here's the code to plot a bar plot and a horizontal bar plot using the given data:
```

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar Plot of Company Profits")
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Horizontal Bar Plot of Company Profits")
plt.show()
```
